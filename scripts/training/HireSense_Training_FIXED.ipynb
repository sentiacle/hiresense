# HireSense Training (Fixed)
This version fixes PDF loading, OCR, and dataset filtering issues.

```
import os
import pytesseract
from pdf2image import convert_from_path
import pdfplumber
from tqdm import tqdm

# IMPORTANT (Windows users)
pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"

data_path = "./data/kaggle_resume_pdf"

pdf_files = []
for root, dirs, files in os.walk(data_path):
    for f in files:
        if f.lower().endswith('.pdf'):
            pdf_files.append(os.path.join(root, f))

print(f"Found {len(pdf_files)} PDFs")
```

```
def extract_text(filepath):
    text = ""

    try:
        with pdfplumber.open(filepath) as pdf:
            for page in pdf.pages:
                t = page.extract_text()
                if t:
                    text += t + "\n"
    except:
        pass

    if len(text.strip()) < 30:
        try:
            images = convert_from_path(filepath, dpi=200, first_page=1, last_page=2)
            ocr_text = ""
            for img in images:
                ocr_text += pytesseract.image_to_string(img) + "\n"
            text = ocr_text
        except:
            pass

    return text.strip()
```

```
texts = []
for f in tqdm(pdf_files):
    t = extract_text(f)
    if len(t) > 20:
        texts.append(t)

print("Loaded:", len(texts))
```
